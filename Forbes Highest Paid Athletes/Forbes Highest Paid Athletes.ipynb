## Import Relevant Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

## Load the Data

```python
data = pd.read_csv("Forbes Richest Atheletes (Forbes Richest Athletes 1990-2020).csv")
data.head()
```

## About the Data

Acronym | Description
--- | ---
S.No | Number
Name | Name of the Athlete
Nationality | Nationality of the Athele
Current Rank | Ranking of the Athele that year
Previous Year Rank | Ranking of the athlete prior to that year
Sport | Sport that the athlete do
Year | Year of the rank
Earnings ($ Million) | Earning in Million 

Note : 2001 didn't have any data

```python
new_data = data.drop('Previous Year Rank',axis=1)
new_data
```

```python
new_data.describe()
```

```python
new_data['Sport'].mask(new_data['Name']=="Sugar Ray Leonard", "Boxing",inplace=True)
```

```python
new_data[new_data['Name']=="Sugar Ray Leonard"]
```

```python
new_data['Sport'] = new_data['Sport'].replace({'basketball':'Basketball',
                                               'NBA':'Basketball',
                                              'boxing' : 'Boxing',
                                              'golf' : 'Golf',
                                              'auto racing' :'F1 Motorsports',
                                              'Auto Racing' : 'F1 Motorsports',
                                              'Auto racing' : 'F1 Motorsports',
                                               'F1 racing' : 'F1 Motorsports',
                                              'motorcycle gp' : 'MotoGP',
                                              'tennis' : 'Tennis',
                                              'NFL': 'American Football',
                                              'Ice Hockey' : 'Hockey',
                                              'ice hockey' : 'Hockey',
                                              'baseball' : 'Baseball',
                                              'soccer' : 'Soccer',
                                              'Auto Racing (Nascar)': 'NASCAR',
                                              'cycling' : 'Cycling'})
```

```python
new_data['Sport'].value_counts()
```

```python
count_sports = new_data['Sport'].value_counts().tolist()
```

```python
sports_name = ['Basketball',
              'Boxing', 'Golf',
              'Soccer','F1 Motorsports',
              'Tennis','American Football',
              'Baseball','NASCAR','Hockey',
              'MotoGP','American Football / Baseball',
              'Cycling','MMA']
```

```python
total_sports = pd.DataFrame()
total_sports['Name'] =sports_name
total_sports['Count of Athletes on the List'] = count_sports
```

```python
total_sports
```

- Basketball is the most sport that appear in this list.
