## Column Explanation

### lter

Acronym | Description
--- | ---
studyName | Study Name of the Penguins
SampleNumber | Sample Number of Penguins
Species | Species of the Penguins 
Region | Region of the Penguins
Island | Island Name (Dream, Torgesen, Biscoe)
Stage | Growth Stage of the Penguins
Individual ID | ID of each Penguins
Clutch Completion | a character string denoting if the study nest observed with a full clutch, i.e., 2 eggs
Date Egg | a date denoting the date study nest observed with 1 egg (sampled)
Culmen Length | Length of the Culmen (mm)
Culmen Depth | Depth of the Culmen (mm)
Flipper Length | Length of the Flipper (mm)
Body Mass | Body Mass of the Penguins (g)
Sex | Male/Female
Delta 15 N | a number denoting the measure of the ratio of stable isotopes 15N:14N
Delta 13 C | a number denoting the measure of the ratio of stable isotopes 13C:12C
Comments | Comments for each Penguins (Note)

### Count from each species on this data

```python
size_edited['species'].value_counts()
```

```python
sns.set(rc={'figure.figsize':(10,10)})
sns.scatterplot(size_edited['culmen_length_mm'],size_edited['culmen_depth_mm'],hue=size_edited['species'])
```

- Adelie : Shortest based on culmen length but has the largest culmen depth (culmen depth similar to chinstrap)
- Chinstrap : 2nd longest culmen length and also 2 largest culmen depth.
- Gentoo : has the smalles culmen depth, but the longest culmen length.

```python
sns.scatterplot(size_edited['flipper_length_mm'],size_edited['body_mass_g'],hue=size_edited['species'])
```

- The point that we can take here, is that based on Flipper Length and Body Mass Adelie and Chinstrap are quite similiar. 
- Gentoo stand out because they have larger flipper length and a heavier body mass, compared to other 2 species.
